## Perceptron

[The perceptron: a probabilistic model for information storage and organization in the brain](https://pubmed.ncbi.nlm.nih.gov/13602029/)

$$
\begin{array}{ccccl}
x_1 & \xrightarrow{\;w_1\;} & \boxed{\begin{array}{c}\text{sum}\\ z=w_1x_1+w_2x_2\\ \text{compare with } \theta\end{array}} & \xrightarrow{\text{step}} & y \\
x_2 & \xrightarrow{\;w_2\;} & & &
\end{array}
$$

$$
y=
\begin{cases}
1, & z > \theta \\
0, & z \le \theta
\end{cases}
$$

<div align="center">

| x1 | x2 | AND | NAND | OR |
|---:|---:|----:|-----:|---:|
| 0 | 0 | 0 | 1 | 0 |
| 0 | 1 | 0 | 1 | 1 |
| 1 | 0 | 0 | 1 | 1 |
| 1 | 1 | 1 | 0 | 1 |

</div>




```python
# Simple Perceptron 구현

X = ((0, 0), (0, 1), (1, 0), (1, 1))

def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1 * w1 + x2 * w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
    
```
